```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('flowerr.jpg')
plt.subplot(2,3,1)
plt.title("Original")
plt.imshow(image)

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.subplot(2,3,2)
plt.title("Grayscale")
plt.imshow(gray_image, cmap = 'gray')

_, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)
plt.subplot(2,3,3)
plt.title("Binary")
plt.imshow(binary_image, cmap = 'gray')

blue_channel, green_channel, red_channel = cv2.split(image)

plt.subplot(2,3,4)
plt.title('Red Channel')
plt.imshow(red_channel, cmap = 'Reds')

plt.subplot(2,3,5)
plt.title('Green Channel')
plt.imshow(green_channel, cmap='Greens')

plt.subplot(2,3,6)
plt.title('Blue Channel')
plt.imshow(blue_channel, cmap='Blues')
```

```python
import cv2
import numpy as np

# Read the image
img = cv2.imread("dog.jpg", cv2.IMREAD_GRAYSCALE)

# Calculate mean and standard deviation
mean_value = np.mean(img)
std_deviation = np.std(img)

# Calculate correlation coefficient
correlation_coefficient = np.corrcoef(img)[0, 1]

# Display results
print("Mean: {:.2f}".format(mean_value))
print("Standard Deviation: {:.2f}".format(std_deviation))
print("Correlation Coefficient: {:.2f}".format(correlation_coefficient))
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
img = cv2.imread("monkey.jpg", cv2.IMREAD_GRAYSCALE)

# Iterate through each bit plane
for i in range(8):
    # Extract the i-th bit plane
    bit_plane = (img >> i) & 1

    # Display the bit plane
    plt.subplot(2, 4, i+1)
    plt.imshow(bit_plane, cmap='gray')
    plt.title(f'Bit Plane {i}')
    plt.axis('off')

# Adjust layout and display
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
img = cv2.imread("ganesh.jpg", cv2.IMREAD_GRAYSCALE)

low_threshold = 100  
high_threshold = 200

# Apply intensity slicing
result_img = np.zeros_like(img)
result_img[(img >= low_threshold) & (img <= high_threshold)] = 255

# Plot the original and result images using Matplotlib
plt.figure(figsize=(10, 5))

# Original image
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Enhanced image
plt.subplot(1, 2, 2)
plt.imshow(result_img, cmap='gray')
plt.title('Enhanced Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
