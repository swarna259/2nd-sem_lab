```python
import cv2

img = cv2.imread("sunflower.jpg", 0)
print("Original matrix\n",img)

rows, cols = img.shape


rowMatrix = []

for i in range(rows):
    for j in range(cols):
        rowMatrix.append(img[i,j])
        
print("Row major representation of matrix\n",rowMatrix)
```

```python
import math

v1 = [3, 6, 1, 11, 10, 5]
v2 = [4, 8, 1, 3, 13, 7]
sum1 = 0
for i in range(len(v1)):
    sum1 += (v1[i] - v2[i]) ** 2
ED = math.sqrt(sum1)

print("Eucidean distance :: ",ED)
```

```python
import cv2

img = cv2.imread("sunflower.jpg", 0)
print("Original matrix\n",img)

rows, cols = img.shape

sum1 = 0

for i in range(rows):
    for j in range(cols):
        sum1 += img[i,j]
mean = sum1 / (rows * cols)
print("Mean :: ", mean)

varianceSum = 0
for i in range(rows):
    for j in range(cols):
        varianceSum += (img[i,j] - mean) ** 2
variance = varianceSum / (rows * cols)
print("Variance :: ", variance)
```

```python
import cv2

A = [[2, 5, 3],
    [6, 1, 8],
    [4, 7, 8]]
print("Original matrix\n",A)

sum1 = 0

for i in range(3):
    for j in range(3):
        sum1 += A[i][j]
mean = sum1 / 9
print("Mean :: ", mean)

varianceSum = 0
for i in range(3):
    for j in range(3):
        varianceSum += (A[i][j] - mean) ** 2
variance = varianceSum / 9
print("Variance :: ", variance)
```
